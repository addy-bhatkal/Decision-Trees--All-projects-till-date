```python
#Importing all the libraries and algorithms

import pandas as pd, numpy as np
import os
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)
```

```python
#Running the correlation heatmap, importing datetime to check time for the algo to run. 
import datetime, time
os.chdir('C:\\Users\\ABC\\6. Datasets')
df= pd.read_csv('mushroom_cleaned.csv')
corr1 = df.corr()
sns.heatmap(corr1, annot =True)
```

```python
#Scaling the Data
mm = MinMaxScaler()
df = pd.DataFrame(mm.fit_transform(df), columns=df.columns)
df.head()
x = df.iloc[:, 1:-1]
y = df['class']
```

```python
#Running the Decision Tree algorithm
start_time = time.strftime("%H:%M:%S")
print(start_time)

dt = DecisionTreeClassifier()
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)

model = dt.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_prob = model.predict_proba(x_test)[:, 1]
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
if acc > 0.95:
    print(acc)
end_time = time.strftime("%H:%M:%S")
print(end_time)
sns.heatmap(cm, annot  = True)
```

```python
#Cross Validating the Training data
cvv1 = cross_val_score(dt, x_train, y_train, cv = 5, scoring='roc_auc')
print('cvv score:', cvv1.mean())
```

```python
#Cross Validating the Testing data
cvv2 = cross_val_score(dt, x_test, y_test, cv = 5, scoring='roc_auc')
print('cvv score:', cvv2.mean())
```
